```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
import numpy as np
```

```python
dfd=pd.read_csv('deliveries.csv')
```

```python
dfd.head()
```

```python
dfm=pd.read_csv('matches.csv')
```

```python
dfm.head()
dfm=dfm[['season','team1','team2','winner','win_by_runs','win_by_wickets','player_of_match','toss_winner','toss_decision']]
```

```python
dfm.head()
```

```python
for cols in dfm.columns:
    print(cols)
    print(dfm[cols].value_counts())
```

```python
dfm.isna().sum()
```

```python
dfm.dropna(inplace=True)
```

```python

fig=plt.gcf()
fig.set_size_inches(25,8)


sns.countplot(x='winner',data=dfm,hue='toss_decision')
```

#MUMBAI INDIANS, CHENNAI SUPER KINGS , KOLKATA KNIGHT RIDERS ARE TOP WINNER

```python
fig=plt.gcf()
fig.set_size_inches(25,8)

sns.barplot(x='winner',y='win_by_runs',data=dfm)
```

#CHENNAI SUPER KING HAS WON MOST OF THE MATCHES BY MAKING RUNS ROYAL CHALLENGERS BANGALORE AND MUMBAI INDIANS HAVE ALSO WON THE MATCHES BY MAKING RUNS

```python
fig=plt.gcf()
fig.set_size_inches(28,8)

sns.barplot(x='winner',y='win_by_wickets',data=dfm)
```

#TEAM LIKE GUJARAT LIONS,KOCHI TUSKERS KERELA WHICH ARE ONE OF WEAKEST TEAMS, THEIR POWERFUL POINT IS BOWLING AND WEAK POINT IS BATING WHILE WINING TEAMS WIN THE MATCHES BY BATING

```python

fig=plt.gcf()
fig.set_size_inches(25,8)


sns.countplot(x='winner',data=dfm,hue='toss_decision')
```

#FROM THE GRAPH WE CAN SEE WHICH TEAMS ARE GOOD IN WHICH PART (BATTING/BOWLING)

```python

fig=plt.gcf()
fig.set_size_inches(25,8)


sns.countplot(x='toss_winner',data=dfm,hue='winner')
```
